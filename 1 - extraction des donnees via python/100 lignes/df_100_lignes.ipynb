```python
import requests
import pandas as pd
from sqlalchemy import create_engine
```

```python
# URL de l'appel API (data.economie.gouv.fr)
url = 'https://data.economie.gouv.fr/api/explore/v2.1/catalog/datasets/prix-des-carburants-en-france-flux-instantane-v2/records?limit=100&offset=0'

# r = reponse du serveur qui heberge l'API (suite à la requete)
r = requests.get(url)

# Verification du statut de la reponse
if r.status_code == 200:
    data = r.json()  # Conversion de la réponse JSON en dictionnaire Python
else:
    print(f"Erreur : {r.status_code}")

# Importation de l'objet JSON en dataframe pandas
df = pd.json_normalize(data['results'])




# Pour afficher toutes les colonnes
pd.set_option('display.max_columns', None)

# Pour afficher toutes les lignes
#pd.set_option('display.max_rows', None)

# Affichage du DataFrame
df

# Exporter le DataFrame en fichier CSV
df.to_csv('carburants_france.csv', index=False)
```
