```python
import requests
import pandas as pd
```

```python
# URL de base de l'appel API
base_url = 'https://data.economie.gouv.fr/api/explore/v2.1/catalog/datasets/prix-des-carburants-en-france-flux-instantane-v2/records'

# Initialiser une liste pour stocker toutes les données
all_data = []
limit = 100  # Limite par requête
offset = 0   # Décalage pour la pagination

while True:
    # Construire l'URL avec les paramètres de limite et d'offset
    url = f"{base_url}?limit={limit}&offset={offset}"

    # Effectuer la requête
    r = requests.get(url)

    # Vérification du statut
    if r.status_code == 200:
        data = r.json()  # Convertir la réponse JSON en dictionnaire Python

        # Ajouter les résultats à la liste
        all_data.extend(data['results'])

        # Vérifier si nous avons récupéré toutes les données
        if len(data['results']) < limit:
            break  # Si le nombre de résultats est inférieur à la limite, on a fini
        else:
            offset += limit  # Augmenter l'offset pour la prochaine requête
    else:
        print(f"Erreur : {r.status_code}")
        break

# Importation de l'objet JSON en dataframe pandas
df = pd.json_normalize(all_data)

# Afficher le DataFrame
display(df)
```
